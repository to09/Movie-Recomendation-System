```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rat = pd.read_csv("ratings.csv")
```

```python
rating = pd.DataFrame(rat,columns=["userId","movieId","rating","timestamp"])
```

```python
rating.head()
```

```python
movie = pd.read_csv("movies.csv")
```

```python
movie = movie.drop("genres",axis=1)
```

```python
df = pd.merge(rating,movie,on="movieId")
```

```python
df.head()
```

# movies with the best rating


```python
df.groupby("title")["rating"].mean().sort_values(ascending=False).head()
```

## title having most rating

```python
df.groupby("title")["rating"].count().sort_values(ascending=False).head()
```

```python
rating = pd.DataFrame(df.groupby("title")["rating"].mean())
```

```python
rating.head()
```

```python
rating["no. of rating"] = df.groupby("title")["rating"].count()
```

```python
rating.head()
```

```python
rating["no. of rating"].hist(bins=30)
```
